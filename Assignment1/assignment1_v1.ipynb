# Task 2: Compute the percentage of capitalized words in spam and ham messages

```python
def isCapital(word):
    if word == "'" or word == "":
        return False
    for i in word:
        if i == "'" and len(word)!=1:
            continue
        if not i.isalpha() or i.islower():
            return False
    return True
```

```python
spamTotal = 0
spamCap = 0
hamTotal = 0
hamCap = 0
for category, message in df.iterrows(): 
    if message['Category'] == 'spam':
        words = re.split("\s|(?<!\d)[\\\,\.\/\!\:\?\(\)\[\]\-\&](?!\d)", message['Message'])
        for word in words:
            if len(word) and isCapital(word):
                spamCap += 1
            if len(word):
                spamTotal += 1
    else:
        words = re.split("\s|(?<!\d)[\\\,\.\/\!\:\?\(\)\[\]\-\&](?!\d)", message['Message'])
        for word in words:
            if len(word) and isCapital(word):
                hamCap += 1
            if len(word):
                hamTotal += 1
```

```python
print("Spam Total: ", spamTotal)
print("Spam Cap: ", spamCap)
print("Total Ham: ", hamTotal)
print("Ham Cap: ", hamCap)
```

```python
percentCapInSpam = spamCap * 100 / spamTotal
percentCapInHam = hamCap * 100 / hamTotal
```

```python
print("Percentage of capitalized words in Spam messages = " + str(percentCapInSpam) + " %")
print("Percentage of capitalized words in Ham messages = " + str(percentCapInHam) + " %")
```

# Task 9: Spam/Ham classification

```python
trueVals = []
predVals = []
for category, message in df.iterrows():
    if message['Category'] == "ham":
        trueVals.append(0)
    else:
        trueVals.append(1)
```

```python
def getAccuracy(trueVals, predVals):
    trueSpam = 0
    falseSpam = 0
    trueHam = 0
    falseHam = 0
    for i in range(len(trueVals)):
        if trueVals[i] == predVals[i]:
            if trueVals[i] == 1:
                trueSpam += 1
            else:
                trueHam += 1
        else:
            if trueVals[i] == 0:
                falseSpam += 1
            else:
                falseHam += 1
    print("True Spam: " + str(trueSpam))
    print("True Ham: " + str(trueHam))
    print("False Spam: " + str(falseSpam))
    print("False Ham: " + str(falseHam))
    print("accuracy = " + str((trueSpam + trueHam) /
                              (trueSpam + trueHam + falseSpam + falseHam)))
```

### Looking at the above data, we can say that the percentage of capitalized words in spam and ham messages are approximately equal. Just for the sake of comparison, we can say that the spam messages are more likely to contan spam messages as compared to ham.
To create a model, we can say that if a message contains capitalized word, it is more likely to bea spam message.

```python
noOfHam = 0
noOfSpam = 0
avgInSpam = []
avgInHam = []
trueVals = []
predVals = []
for category, message in df.iterrows():
    numberOfCaps = 0
    numberOfWords = 0
    words = re.split("\s|(?<!\d)[,./!:?\(\)\[\]\-\&](?!\d)", message['Message'])
    for word in words:
        if isCapital(word):
            numberOfCaps += 1
        numberOfWords += 1
    if message['Category'] == "ham":
        noOfHam += 1
        avgInHam.append(numberOfCaps/numberOfWords)
        trueVals.append(0)
    else:
        noOfSpam += 1
        avgInSpam.append(numberOfCaps/numberOfWords)
        trueVals.append(1)
print(noOfHam, noOfSpam, sum(avgInSpam), sum(avgInHam), len(avgInSpam), len(avgInHam))
averageNumberOfCapsInSpam = sum(avgInSpam)/len(avgInSpam)
averageNumberOfCapsInHam = sum(avgInHam)/len(avgInHam)
print("Average number of capital words in Spam = " + str(averageNumberOfCapsInSpam))
print("Average number of capital words in Ham = " + str(averageNumberOfCapsInHam))
```

```python
def predictBasedOnTask2(message, sep):
    words = re.split("\s|(?<!\d)[,./!:?\(\)\[\]\-\&](?!\d)", message)
    numberOfCaps = 0
    numberOfWords = 0
    for word in words:
        if isCapital(word):
            numberOfCaps += 1
        if word != "":
            numberOfWords += 1
    if numberOfWords == 0:
        return 1
    if numberOfCaps/numberOfWords < sep:
        return 0
    else:
        return 1
```

```python
predVals = []
def predictTask2(sep):
    for category, message in df.iterrows():
        predVals.append(predictBasedOnTask2(message['Message'], sep))
    # getAccuracy(trueVals, predVals)

i = 0.058
while i <= 0.097:
    predVals = []
    i += .001
    print(i)
    predictTask2(i)
    getAccuracy(trueVals, predVals)
    print()
```

Considering the above pattern, where as true spam decreases, true ham start increasing and since the number of true ham is much more than that of true spam, it increases the accuracy. Since the data is not iid, therefore, we consider the value of 0.078 as the dividing value. This means, if the ratio of capital words to all the words is less that 0.078, then it is a ham message, else spam.

```python
sep = 0.078
predVals = []
for category, message in df.iterrows():
    predVals.append(predictBasedOnTask2(message['Message'], sep))

getAccuracy(trueVals, predVals)
```

### If we look at the data of the emails in the messages, it is clear that the emails are more likely to be spam than ham. This however is not a stron metric as the number of messages containing emails is very small(7 in this case, out of which 1 is false positive).


If we try to merge the above two informations, we can see that if a message contains either an email address or a phone number, it is more likely to be a spam message. So with this information, we can classify whether a message is spam, but not the other way around. Still for the sake of it, let us say that any message containing a number or an email address is spam and any message which neither contains a number nor an email address is ham.

```python
def predictBasedOnTask3(message):
    numberList = re.findall(
        r"\b[0]?\d{3}[\s-]?\d{3}[\s-]?\d{4}\b|\b[0]?\d{3}[\s-]?\d{4}[\s-]?\d{3}\b|\b[0]?\d{5}[\s-]?\d{5}\b|\b[0]?\d{4}[\s-]?\d{3}[\s-]?\d{3}\b", message)
    emailList = re.findall(
        r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[a-zA-Z]{2,}\b", message)
    if len(numberList) > 0 or len(emailList) > 0:
        return 1
    else:
        return 0
```
