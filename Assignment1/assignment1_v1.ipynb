# Task 2: Compute the percentage of capitalized words in spam and ham messages

```python
spamSmall = 0
spamCap = 0
hamSmall = 0
hamCap = 0
```

```python
df.head()
```

```python
spamSmall = 0
spamCap = 0
hamSmall = 0
hamCap = 0
for category, message in df.iterrows(): 
    if message['Category'] == 'spam':
        words = re.split("\s|(?<!\d)[,./!:?\(\)\[\]-](?!\d)", message['Message'])
        for word in words:
            if len(word) and word[0].isalpha() and word[0].islower():
                spamSmall += 1
            elif len(word) and word[0].isalpha() and word[0].isupper():
                spamCap += 1
    else:
        words = re.split("\s|(?<!\d)[,./!:?\(\)\[\]-](?!\d)", message['Message'])
        for word in words:
            if len(word) and word[0].isalpha() and word[0].islower():
                hamSmall += 1
            elif len(word) and word[0].isalpha() and word[0].isupper():
                hamCap += 1
```

```python
print("Spam Small: ", spamSmall)
print("Spam Cap: ", spamCap)
print("Ham Small: ", hamSmall)
print("Ham Cap: ", hamCap)
```

```python
print("Percentage of capitalized words in Spam messages = " + str(((spamCap / (spamSmall + spamCap))*100)) + "%")
print("Percentage of capitalized words in Ham messages = " + str(((hamCap / (hamSmall + hamCap))*100)) + "%")
```

```python
emailList = {}
emailInSpam = 0
emailInHam = 0
for category, message in df.iterrows():
    tempList = re.findall(
        r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[a-z]{2,}\b", message['Message'])
    if len(tempList)>0:
        if message['Category'] == 'spam':
            emailInSpam += 1
        else:
            emailInHam += 1
    for email in tempList:
        if email in emailList:
            emailList[email] += 1
        else:
            emailList[email] = 1
for email in emailList:
    print(email + ":" + str(emailList[email]))
print("Email in spam: " + str(emailInSpam))
print("Email in ham: " + str(emailInHam))
print("Percentage of Email in spam: " + str(emailInSpam*100/numberOfSpam) + "%")
print("Percentage of Email in ham: " + str(emailInHam*100/numberOfHam) + "%")
```

```python
numbersList = {}
numberInSpam = 0
numberInHam = 0
for category, message in df.iterrows():
    numberList = re.findall(r"\b[0]?\d{3}[\s-]?\d{3}[\s-]?\d{4}\b", message['Message'])
    for number in numberList:
        if number in numbersList:
            numbersList[number] += 1
        else:
            numbersList[number] = 1
    if len(numberList) > 0:
        if message['Category'] == 'spam':
            numberInSpam += 1
        else:
            numberInHam += 1
for number in numbersList:
    print(number + ":" + str(numbersList[number]))
print("Numbers in spam: " + str(numberInSpam))
print("Numbers in ham: " + str(numberInHam))
print("Percentage of Numbers in spam: " + str(numberInSpam*100/numberOfSpam) + "%")
print("Percentage of Numbers in ham: " + str(numberInHam*100/numberOfHam) + "%")
```
